```python
import pandas as pd
import numpy as np

# Načtení datasetu
df = pd.read_csv("rohlik.csv")

# Kontrola dat
df.head()
```

```python
# Základní informace o datasetu
df.info()
```

```python
# Odstraníme "Kč" a nahradíme desetinnou čárku za tečku
df["Cena_numeric"] = df["Cena"].str.replace(" Kč", "").str.replace(",", ".").astype(float)

# Základní statistika cen
df["Cena_numeric"].describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.barh(df["Název"], df["Cena_numeric"], color='skyblue')
plt.xlabel("Cena (Kč)")
plt.title("Ceny produktů")
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
df.to_csv("produkty_upravene.csv", index=False, encoding="utf-8")
```
